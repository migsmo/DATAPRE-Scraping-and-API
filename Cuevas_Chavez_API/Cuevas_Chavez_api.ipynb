```python
import requests
import pandas as pd
import json
from newsapi import NewsApiClient
newsapi = NewsApiClient(api_key='4455f7b7edbe4b0f92931e243ca574ac')
```

# Limitations
- x in range 5 because we are only allowed to get up to 5 pages of data (100 articles)
- date is from 2022-05-0 to 2022-05-03 because we are limited to articles only within the month
- content column which is synonymous to "full article" is only limited to 200 characters

```python
article_data = []
for x in range(5):
    all_articles = newsapi.get_everything(sources='abc-news',
                                      from_param='2022-05-01',
                                      to='2022-05-03',
                                      language='en',
                                      page=x+1)['articles']
    for y in range(len(all_articles)): 
        article_data.append(all_articles[y])
```

```python
article_data_df = pd.DataFrame(article_data)

required_data = article_data_df[['publishedAt', 'title', 'author', 'content']]
required_data
```

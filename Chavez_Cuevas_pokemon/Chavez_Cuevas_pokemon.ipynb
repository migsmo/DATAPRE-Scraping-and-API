## Import statements

```python
import requests
import pandas as pd
import datetime
URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
```

## Setting up Beautiful Soup

```python
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

## Scraping Pokemon data from Bulbapedia

```python
poke_content=soup.find(id='mw-content-text')
poke_tables=poke_content.find_all('table')
```

```python
poke_tables[1]
```

```python
poke_info = []

for i, poke_table in enumerate(poke_tables):
    if i == 0:
        continue;
    else: 
        poke_tbody = poke_table.find('tbody')
        for j, rows in enumerate(poke_tbody.find_all('tr')):
            if j == 0:
                continue
            else:
                try:
                    ndex = rows.find_all('td')[0].text.strip('\n')

                    kdex = rows.find_all('td')[1].text.strip('\n')

                    poke_name = rows.find_all('td')[2].text.strip('\n')

                    poke_type1 = rows.find_all('td')[3].text.strip('\n')

                    poke_type2 = rows.find_all('td')[4].text.strip('\n')
                    
                    poke_generation = i
                    
                    scrape_time = datetime.datetime.now()

                    poke_info.append((kdex, ndex, poke_name, poke_type1, poke_type2, poke_generation, scrape_time))
                except:
                    poke_generation = i
                    
                    scrape_time = datetime.datetime.now()

                    poke_type2 = ''
                    poke_info.append((kdex, ndex, poke_name, poke_type1, poke_type2, poke_generation, scrape_time))
                    
for i in poke_info:
    print(i)
```

## Converting data into Pandas Dataframe

```python
df_pokemon = pd.DataFrame(poke_info)
```

```python
df_pokemon.columns = ['Kdex', 'Ndex', 'Pokemon', 'Type 1', 'Type2', 'Generation', 'Scrapetime']
```

```python
df_pokemon.drop(df_pokemon.index[df_pokemon['Generation'] > 9], inplace=True)
```
